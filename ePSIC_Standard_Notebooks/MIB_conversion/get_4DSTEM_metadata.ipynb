```python
import glob
import pandas as pd
import os
from pathlib import Path
import h5py

##todo : put these functions into ePSIC tools:

def get_allfiles_metadata(ignore_list = []):
    #run this from within your processing directory
    #ignore_list : list of strings containing subdirectories to ignore **INCLUDE PTYCHO RECON SUBDIR HERE**
    #returns a pandas_data_frame 
    #TODO - auto save ?
    
    path_processing = os.getcwd()
    meta_data_list = glob.glob(path_processing + '/**/*.hdf', recursive = True)
    
    if len(ignore_list) != 0:
        for this_string in ignore_list:
            meta_data_list = [p for p in meta_data_list if this_string not in Path(p).parts]

    return(hdf5_list_to_dataframe(meta_data_list))

def extract_hdf5_key_values(fp):
    """Return a flat dict of all key:value pairs in one HDF5 file."""
    meta_data = {}

    with h5py.File(fp, "r") as f:
        for key in f["metadata"].keys():
            try:
                meta_data[key] = f["metadata"][key][()]
            except Exception:
                meta_data[key] = " ***** "

    return meta_data


def hdf5_list_to_dataframe(file_list):
    rows = []

    for fp in file_list:
        fp = Path(fp)
        print(fp)

        metadata = extract_hdf5_key_values(fp)

        row = {
            "source_folder": fp.parent.parent.name,
            "source_file": fp.name,
            **metadata
        }

        rows.append(row)

    return pd.DataFrame(rows)
```

```python
ignore_list = ['pty_out' , 'simulations']
df = get_allfiles_metadata(ignore_list = ignore_list)
```
